```python
# view dataset
col_names = ['fLength', 'fWidth', 'fSize', 'fConc', 'fConc1',
             'fAsym', 'fM3Long', 'fM3Trans', 'fAlpha', 'fDist',
             'class'] # set column names in dataset
df = pd.read_csv("magic04.data", names=col_names) # read data
df.head() # display first five columns
```

```python
# view class component
df['class'].unique()
# classes are g for gamma and h for hadron
```

```python
# convert class labels into 1 and 0. 
# '1' for gamma, '0' for hadron
df['class'] = (df['class'] == 'g').astype(int)
```

```python
df
```

```python
# plot a histogram to visualize features of dataset
for label in col_names[:-1]:
  plt.hist(df[df['class']==1][label], label='gamma', color='green', alpha=0.7, density=True)
  plt.hist(df[df['class']==0][label], label='hadron', color='red', alpha=0.7, density=True)
  plt.title(label)
  plt.ylabel("Probability")
  plt.xlabel(label)
  plt.legend()
  plt.show()
```
